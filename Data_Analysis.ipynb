```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('classes.csv',header=None)
df
```

```python
counts = df.groupby(1).count().sort_values(0,ascending=False).reset_index().rename(columns={1:'class',0:'frequency'})[:20]
plt.figure(figsize=(10,6))
sns.barplot(data=counts,x='class',y='frequency',order=counts['class'],color='#679efc')
plt.title('Top 20 Classes by Frequency')
```
