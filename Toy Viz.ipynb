```python
from sklearn.datasets import make_blobs
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set()
%matplotlib inline
```

```python
blobs, labels = make_blobs(centers=4)
data = pd.DataFrame(blobs,columns=['x','y'])
data['clusters'] = labels
data
```

```python
data['class'] = [np.random.choice(['dog','dog','dog','cat','cat','fox']) if x==3
                else np.random.choice(['burger','sandwich','sub','sandwich','sub','sandwich']) if x==2
                else np.random.choice(['car','car','car','car','truck','van','van']) if x==1
                else np.random.choice(['eagle','pigeon','pigeon','hawk']) for x in data['clusters']]
```

```python
sns.scatterplot(data=data,x='x',y='y',hue='clusters',legend='full',palette='Dark2')
plt.title('Clusters of Images at Convolution Layer $i$')
```

```python
sns.barplot(data=data.loc[data['clusters']==0]['class'].value_counts().reset_index(),x='index',y='class')
plt.ylabel('frequency')
plt.xlabel('class')
plt.title('Distribution of True Labels in Cluster 0')
```
